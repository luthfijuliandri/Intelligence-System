Import library pandas dan mlxtend untuk penggunaan apriori dan aturan asosiasi

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

Definisikan dataset menggunakan array berdasarkan soal yang diberikan

```python
dataset = [['Telur', 'Beras'],
           ['Telur', 'Beras', 'Minyak', 'Topi'],
           ['Telur', 'Beras', 'Minyak', 'Buku'],
           ['Buku', 'Minyak', 'Topi'],
           ['Telur', 'Beras', 'Minyak', 'Topi'],
           ['Telur', 'Beras', 'Minyak'],
           ['Telur', 'Minyak', 'Buku'],
           ['Beras', 'Buku'],
         ]
```

```python
dataset
```

Memakai TransactionEncoder agar array dataset di convert ke dataframe

```python
te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df
```

# **Implementasi Apriori**

Menggunakan library apriori untuk melakukan perhitungan frequent item set. Nilai minimum (ɸ = 2), Sehingga min supportnya adalah **2/8 = 0.25**

```python
frequent_itemsets = apriori(df, min_support=0.25, use_colnames=True)
```

```python
frequent_itemsets
```

*Dilakukan sampai dengan F4. Dengan demikian F4 adalah (Beras, Telur, Minyak, Topi) karena hanya kombinasi inilah yang memenuhi syarat minimal support yang ditentukan.*

## **Implementasi Aturan Asosiasi**

Menggunakan library association_rules untuk melakukan perhitungan aturan asosiasi dengan confidence minimum 60% atau 0.6.

```python
res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```

```python
res
```

Hasil dari association rule di rapikan sehingga kolom yang muncul hanya antecendents consequent support confidence dan lift.

### Aturan Asosisasi Final

Mencari nilai akhir dengan mengkalikan support dan confidence.

```python
res['supportxconfidence'] = res['support'] * res['confidence']
res_sort= res[['antecedents', 'consequents', 'support', 'confidence', 'supportxconfidence']]

res_sort
```

### Hasil Lift Ratio

Aturan yang digunakan adalah lift ratio >= 1 sehingga melakukan pemangkasan aturan yang dibawah 1

```python
res_sort = res[res['lift'] >= 1]
res_final = res_sort[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
res_final
```

# Kesimpulan

Didapatkan 34 Aturan setelah dilakukan aturan asosisasi dimana confidence = 60% dan lift_ratio >= 1. Sehingga Hasil akhir yang didapatkan adalah sebagai berikut!

```python
res_final = res[['antecedents', 'consequents', 'supportxconfidence']]
res_final = res_final.sort_values(by='supportxconfidence', ascending=False)
res_final
```
